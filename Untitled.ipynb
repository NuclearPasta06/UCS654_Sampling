```python
import pandas as pd
import numpy as np
from pycaret.classification import *
```

```python
data = pd.read_csv("Creditcard_data.csv")
data
```

```python
print((data < 0).sum().sum())
print((data > 0).sum().sum())
```

```python
sum(data['Class'])
from sklearn.preprocessing import normalize
Amount = normalize([data['Amount']])[0]
data['Amount'] = Amount
```

```python
X = data.drop(['Class','Time'], axis=1)
y = data['Class']
```

```python
X.head()
```

# ***Oversampling***

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
print(y.value_counts())
```

```python
new_df = pd.concat([X, y], axis=1)
new_df
```

# Sampling:

## ***1. Simple Random Sampling***

```python
# simple random sampling
n = int((2*2 * 0.5*0.5)/(0.05**2))
sample_df = new_df.sample(n=n, random_state=42)
sample_df
```

```python
setup(data = sample_df, target='Class')
cm = compare_models()
```

## ***2. Stratified sampling***

```python
# stratified sampling

n = int((2*2*0.5*0.5)/((0.05)**2))

strata = new_df.groupby('Class')

# sample 2 rows from each stratum
stratified_df = strata.apply(lambda x: x.sample(n))

stratified_df
```

```python
setup(data = stratified_df, target='Class')
cm = compare_models()
```

## ***3. Systematic Sampling***

```python
# systematic sampling
interval = 2
systematic_df = new_df.iloc[::interval]
systematic_df
```

```python
setup(data = stratified_df, target='Class')
cm = compare_models()
```

## ***4. Cluster Sampling***

```python
# Cluster sampling
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=20, random_state=42).fit(new_df)
cluster_assignments = kmeans.labels_

# Select the clusters you want to include in the sample
selected_clusters = [0, 2, 4, 5, 8, 10, 15, 19]

cluster_series = pd.Series(cluster_assignments)

# Create the new DataFrame containing only the rows from the selected clusters
df_cluster_sample = new_df[cluster_series.isin(selected_clusters)]


# Print the resulting DataFrame
(df_cluster_sample)
```

```python
setup(data = stratified_df, target='Class')
cm = compare_models()
```

## ***5. Convenience Sampling***

```python
convenience_sample = pd.concat([new_df.head(330), new_df.tail(330)])
convenience_sample
```

```python
setup(data = convenience_sample, target='Class')
cm = compare_models()
```
